```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline
```

```python
pokemon = pd.read_csv('Data/pokemon.csv')
pokemon
```

```python
pokemon.shape
```

```python
pokemon.head(10)
```

```python
sb.countplot(data = pokemon,x = 'type_1');
```

```python
ax = sb.countplot(data = pokemon,x = 'type_1');
ax.set_xticklabels(ax.get_xticklabels(),rotation=50);
```

```python
sb.countplot(data = pokemon,y = 'type_1');
```

```python
sb.countplot(data = pokemon,x = 'generation_id');
```

```python
gen_order = pokemon['generation_id'].value_counts().index
print(gen_order)
```

```python
sb.countplot(data=pokemon,x='generation_id',order=gen_order);
```

```python
base_color = sb.color_palette()[0]
```

```python
sb.countplot(data=pokemon,x='generation_id',order=gen_order,color=base_color);
```
